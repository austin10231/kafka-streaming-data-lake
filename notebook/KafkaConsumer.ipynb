```python
import pandas as pd
from kafka import KafkaConsumer, KafkaProducer
from time import sleep
from json import dumps, loads
import json
from s3fs import S3FileSystem
```

```python
consumer = KafkaConsumer('demo_test', bootstrap_servers=['52.53.227.127:9092'],
                        value_deserializer=lambda x: loads(x.decode('utf-8')))
```

```python
# for c in consumer:
#     print(c.value)
```

```python
s3 = S3FileSystem()
```

```python
for count, i in enumerate(consumer):
    with s3.open('s3://kafka-stock-market-project-he/stock_market_{}.json'.format(count), 'w')as file:
        json.dump(i.value, file)
```
